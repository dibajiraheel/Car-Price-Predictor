# Hyper parameter tuning for Random Forest Regressor

```python
tf1_rfr = ColumnTransformer([('OHE' , OneHotEncoder(sparse=False , drop='first') , [8,9])] , remainder = 'passthrough')
tf2_rfr = ColumnTransformer([('Ordinal Encoding' , OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = np.nan) , [5,6,8,9,10])] , remainder = 'passthrough')
tf3_rfr = ColumnTransformer([('Nan Imputer' , SimpleImputer(missing_values = np.nan , strategy = 'most_frequent') , slice(0,11))])
tf4_rfr = ColumnTransformer([('Standard Scaler', StandardScaler(), slice(0,11))])
m_rfr   = RandomForestRegressor()


pipe_rfr = Pipeline([('One Hot Encoder',tf1_rfr),('Ordinal Encoding',tf2_rfr),('Nan Imputer',tf3_rfr),('StandardScaler',tf4_rfr),('Random Forest Regressor',m_rfr)])

# define the grid of hyperparameters to search over
param_grid = {
    'Random Forest Regressor__n_estimators': [50, 100, 150],
    'Random Forest Regressor__max_depth': [5, 10, 15],
    'Random Forest Regressor__min_samples_leaf': [1, 2, 4]
}

# perform grid search with cross-validation
grid_search = GridSearchCV(pipe_rfr, param_grid, cv=5, verbose=2, n_jobs=-1)
grid_search.fit(x_train, y_train)

# retrieve the best estimator and make predictions
best_estimator = grid_search.best_estimator_
y_predict_rfr = best_estimator.predict(x_test)

# evaluate the model performance
print("MAE = ",mean_absolute_error(y_test,y_predict_rfr))
print("MSE = ",mean_squared_error(y_test,y_predict_rfr))
print("RMSE = ",np.sqrt(mean_squared_error(y_test,y_predict_rfr)))
print('r2 Score = ',r2_score(y_test,y_predict_rfr))

set_config(display='diagram')
best_estimator
```

# Lets try Artificial Neural Networks ( ANN )

```python
# Define the function to create the ANN model
def create_ann_model(hidden_layers=1, hidden_units=32, dropout_rate=0.2, optimizer='adam'):
    model = keras.Sequential()
    model.add(keras.layers.Dense(units=hidden_units, input_dim=11, activation='relu'))
    for i in range(hidden_layers):
        model.add(keras.layers.Dense(units=hidden_units, activation='relu'))
        model.add(keras.layers.Dropout(rate=dropout_rate))
    model.add(keras.layers.Dense(units=1, activation='linear'))
    model.compile(loss='mean_squared_error', optimizer=optimizer)
    return model

# Define the transformers and pipeline
tf1_ann = ColumnTransformer([('OHE' , OneHotEncoder(sparse=False , drop='first') , [8,9])] , remainder = 'passthrough')
tf2_ann = ColumnTransformer([('Ordinal Encoding' , OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = np.nan) , [5,6,8,9,10])] , remainder = 'passthrough')
tf3_ann = ColumnTransformer([('Nan Imputer' , SimpleImputer(missing_values = np.nan , strategy = 'most_frequent') , slice(0,11))])
tf4_ann = ColumnTransformer([('Standard Scaler', StandardScaler(), slice(0,11))])
pipe_ann = Pipeline([('One Hot Encoder',tf1_ann),('Ordinal Encoding',tf2_ann),('Nan Imputer',tf3_ann),('StandardScaler',tf4_ann)])

# Transform the data with the pipeline
x_train_ann = pipe_ann.fit_transform(x_train)
x_test_ann = pipe_ann.transform(x_test)

# Create the KerasRegressor object for GridSearchCV
ann_model = KerasRegressor(build_fn=create_ann_model, verbose=0)

# Define the hyperparameters to be tuned with their respective values
param_grid = {'hidden_layers': [1, 2, 3, 6, 9, 12], 'hidden_units': [16, 32, 64], 'dropout_rate': [0.1, 0.2, 0.3], 'optimizer': ['adam', 'rmsprop']}

# Perform GridSearchCV with the defined hyperparameters
grid = GridSearchCV(estimator=ann_model, param_grid=param_grid, cv=5, n_jobs=-1)
grid_result = grid.fit(x_train_ann, y_train)

# Predict the values on the test set
y_predict_ann = grid_result.predict(x_test_ann)

# Print the evaluation metrics
print("MAE = ",mean_absolute_error(y_test,y_predict_ann))
print("MSE = ",mean_squared_error(y_test,y_predict_ann))
print("RMSE = ",np.sqrt(mean_squared_error(y_test,y_predict_ann)))
print('r2 Score = ',r2_score(y_test,y_predict_ann))

# Print the pipeline diagram
set_config(display='diagram')
pipe_ann
```
